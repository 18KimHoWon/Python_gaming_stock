```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

# 웹드라이버 설정
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 네이버 금융 페이지 로드

#페이지 설정
page = [i for i in range(1, 5, 1)]

url = 'https://finance.naver.com/item/frgn.naver?code=225570&page='

all_table = pd.DataFrame()  # 빈 데이터프레임 생성

for i in page:
    driver.get(url + str(i))

    # 페이지가 완전히 로드될 때까지 대기
    time.sleep(5)

    # 페이지의 HTML 가져오기
    html = driver.page_source

    # pandas를 사용해 HTML 분석
    tables = pd.read_html(html)

    # print(tables[2])
    table = tables[2].dropna(how='all').dropna(axis=1, how='all')

    all_table = pd.concat([all_table, table])

# 웹드라이버 종료
driver.quit()

# 인덱스 재설정
all_table.reset_index(drop=True, inplace=True)

print(all_table)
```
